# Odds and Log-Odds Demo

I will explain and visualize **odds** and **log-odds** using a simple binary outcome example.

---

## What are Odds?

For a binary event with probability $p$:

$
\text{Odds} = \frac{p}{1-p}
$

- Example: if a coin comes up heads 75% of the time ($p=0.75$), then:
$
\text{Odds} = \frac{0.75}{0.25} = 3
$

- Interpretation: "3 to 1" in favor of heads.

---

## What are Log-Odds?

Log-odds (or **logit**) is just the natural log of odds:

$
\text{Log-Odds} = \log \frac{p}{1-p}
$

- Example:  
$\log(3) \approx 1.0986$  
- Converts probability (0â€“1) to real numbers ($-\infty, +\infty$), useful for regression.

---

## Python Demonstration


```python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0.01, 0.99, 100)
odds = p / (1-p)

log_odds = np.log(odds)

plt.figure(figsize=(8,5))
plt.plot(p, odds, label='Odds', color='blue')
plt.plot(p, log_odds, label='Log-Odds', color='red')
plt.xlabel('Probability p')
plt.ylabel('Value')
plt.title('Odds and Log-Odds vs Probability')
plt.legend()
plt.grid(True)
plt.show()
```

## Why Log-Odds?

We use **log-odds** because it makes the scale **symmetrical** around 0:

* Probability $p$ ranges from 0 to 1.
* Odds $\frac{p}{1-p}$ ranges from 0 to $\infty$ â†’ **not symmetrical**.
* Log-odds $\log\frac{p}{1-p}$ ranges from $-\infty$ to $+\infty$ â†’ **symmetrical** around 0.

This symmetry makes it easier to model increases or decreases in probability **linearly** using regression.
